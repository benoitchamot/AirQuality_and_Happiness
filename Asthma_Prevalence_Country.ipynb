```python
import pandas as pd

# Replace 'filename.csv' with the actual path and filename of your CSV file
csv_file = 'Raw_Datasets/Asthma_Prevalence_By_Country.csv'

# Use pandas to read the CSV file into a DataFrame
asthma_prevalence_df = pd.read_csv(csv_file)

# Display the first few rows of the DataFrame
asthma_prevalence_df.head()
```

```python
#delete the 'Code' column
del asthma_prevalence_df['Code']
asthma_prevalence_df.head()
```

```python
#rename Asthma cases column
asthma_prevalence_df = asthma_prevalence_df.rename(columns={'Current number of cases in the population per 100 people - Sex: Both - Age: Age-standardized - Cause: Asthma': 'Asthma Cases per 100 - Age-Std - Both Sex'})
asthma_prevalence_df.head()
```

```python
#remove and Year values earlier than 2015
asthma_prevalence_df = asthma_prevalence_df[asthma_prevalence_df['Year'] >= 2015]
asthma_prevalence_df.head()
```

```python
#rename 'Entity' column to 'Country'
asthma_prevalence_df = asthma_prevalence_df.rename(columns={'Entity': 'Country'})
asthma_prevalence_df.head()
```

```python
cleaned_asthma_prevalence = 'cleaned_asthma_prevalence.csv'

#export the DataFrame as a CSV file
asthma_prevalence_df.to_csv(cleaned_asthma_prevalence, index=False)
```
