```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# File to Load
life_expectancy_load = Path("Data_Sources/World Life Expectancy.csv")
```

```python
# Read Data File and store into Pandas DataFrames
life_expectancy_data = pd.read_csv(life_expectancy_load)
life_expectancy_data
```

```python
#Get the header columns
life_expectancy_data.columns.values.tolist()
```

```python
# Drop unwanted columns and last empty column
drop_columns = life_expectancy_data[3:].drop(["Unnamed: 2", "Unnamed: 3",  'Unnamed: 2',
 'Unnamed: 3',
 'Unnamed: 4',
 'Unnamed: 5',
 'Unnamed: 6',
 'Unnamed: 7',
 'Unnamed: 8',
 'Unnamed: 9',
 'Unnamed: 10',
 'Unnamed: 11',
 'Unnamed: 12',
 'Unnamed: 13',
 'Unnamed: 14',
 'Unnamed: 15',
 'Unnamed: 16',
 'Unnamed: 17',
 'Unnamed: 18',
 'Unnamed: 19',
 'Unnamed: 20',
 'Unnamed: 21',
 'Unnamed: 22',
 'Unnamed: 23',
 'Unnamed: 24',
 'Unnamed: 25',
 'Unnamed: 26',
 'Unnamed: 27',
 'Unnamed: 28',
 'Unnamed: 29',
 'Unnamed: 30',
 'Unnamed: 31',
 'Unnamed: 32',
 'Unnamed: 33',
 'Unnamed: 34',
 'Unnamed: 35',
 'Unnamed: 36',
 'Unnamed: 37',
 'Unnamed: 38',
 'Unnamed: 39',
 'Unnamed: 40',
 'Unnamed: 41',
 'Unnamed: 42',
 'Unnamed: 43',
 'Unnamed: 44',
 'Unnamed: 45',
 'Unnamed: 46',
 'Unnamed: 47',
 'Unnamed: 48',
 'Unnamed: 49',
 'Unnamed: 50',
 'Unnamed: 51',
 'Unnamed: 52',
 'Unnamed: 53'], axis=1)
drop_columns.columns = drop_columns.iloc[0]
drop_last_column = drop_columns[1:].iloc[:, :-1].round(2)
drop_last_column
```

```python
# Rename the columns and remove decimal points
drop_last_column.rename(columns=lambda x: str(int(float(x))) if isinstance(x, float) else x, inplace=True)
drop_last_column.fillna(0, inplace=True)
drop_last_column
```

```python
value_vars = [str(year) for year in range(2015, 2021)]
melted_data = pd.melt(drop_last_column, id_vars="Country Name", value_vars=value_vars)
sorted_data = melted_data.sort_values(by=["Country Name", 3], ascending=[True, True])

sorted_data = sorted_data.rename(columns={"Country Name": "Country", 3: "Year", "value": "Life Expectancy"})

sorted_data
```

```python
# Storing file in Data folder
cleaned_data_output_path = Path("Cleaned_Datasets/cleaned_life_expectancy_datav1.csv")
sorted_data.to_csv(cleaned_data_output_path, index=False)
```
