```python
import pandas as pd

# Replace 'filename.csv' with the actual path and filename of your CSV file
csv_file = 'Raw_Datasets/Asthma_Hospitalizations_per_year_and_state_in_USA_2015-2021).csv'

# Use pandas to read the CSV file into a DataFrame
hospital_df = pd.read_csv(csv_file)

# Display the first few rows of the DataFrame
hospital_df.head()
```

```python
#delete columns 'StateFIPS', 'Data Comment'. 'Unnamed:5'
hospital_df.drop(['StateFIPS', 'Data Comment', 'Unnamed: 5'], axis=1, inplace=True)
hospital_df.head()
```

```python
#rename Value column
hospital_df = hospital_df.rename(columns={'Value': 'No. of Hospitalisations'})
hospital_df.head()
```

```python
hospital_df
```

```python
cleaned_hospital_df = 'cleaned_hospitalisations.csv'

#export the DataFrame as a CSV file
hospital_df.to_csv(cleaned_hospital_df, index=False)
```
