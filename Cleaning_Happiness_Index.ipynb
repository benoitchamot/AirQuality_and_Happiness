Data on Happiness Index for years 2015-2019

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
# Study data files
happiness_2015_path = "Data_Sources/Happiness 2015.csv"
happiness_2016_path = "Data_Sources/Happiness 2016.csv"
happiness_2017_path = "Data_Sources/Happiness 2017.csv"
happiness_2018_path = "Data_Sources/Happiness 2018.csv"
happiness_2019_path = "Data_Sources/Happiness 2019.csv"

# Read the happiness data
happiness_2015 = pd.read_csv(happiness_2015_path)
happiness_2016 = pd.read_csv(happiness_2016_path)
happiness_2017 = pd.read_csv(happiness_2017_path)
happiness_2018 = pd.read_csv(happiness_2018_path)
happiness_2019 = pd.read_csv(happiness_2019_path)
```

```python
happiness_2015.head()
```

```python
# Remove columns and only keep countries, year, ranking, and score.
columns_to_drop_1 = ['Region', 'Standard Error', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 
                     'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']
happiness_2015 = happiness_2015.drop(columns_to_drop_1, axis=1)
happiness_2015.head()
```

```python
happiness_2016.head()
```

```python
# Remove columns and only keep countries, year, ranking, and score.
columns_to_drop_2 = ['Region', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Economy (GDP per Capita)', 
                     'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']
happiness_2016 = happiness_2016.drop(columns_to_drop_2, axis=1)
happiness_2016.head()
```

```python
happiness_2017.head()
```

```python
# Remove columns and only keep countries, year, ranking, and score.
columns_to_drop_3 = ['Whisker.high', 'Whisker.low', 'Economy..GDP.per.Capita.',
                     'Family', 'Health..Life.Expectancy.', 'Freedom', 'Generosity', 'Trust..Government.Corruption.', 'Dystopia.Residual']
happiness_2017 = happiness_2017.drop(columns_to_drop_3, axis=1)
happiness_2017.head()
```

```python
happiness_2018.head()
```

```python
# Remove columns and only keep countries, year, ranking, and score.
columns_to_drop_4 = ['GDP per capita', 'Social support', 'Healthy life expectancy',
                    'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']
happiness_2018 = happiness_2018.drop(columns_to_drop_4, axis=1)
```

```python
#Renaming 'country or region' column to 'Country' 
happiness_2018 = happiness_2018.rename(columns={'Country or region': 'Country'})
happiness_2018.head()
```

```python
happiness_2019.head()
```

```python
columns_to_drop_5 = ['GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']
happiness_2019 = happiness_2019.drop(columns_to_drop_5, axis=1)
happiness_2019.head()
```

```python
#Renaming 'country or region' column to 'Country' 
happiness_2019 = happiness_2019.rename(columns={'Country or region': 'Country'})
happiness_2019.head()
```

```python
# Combine the data into a single DataFrame
combined_dataframe1 = pd.merge(happiness_2015, happiness_2016,
                              how='left', on='Country')
```

```python
# check the merged dataframe
print(combined_dataframe1)
```

```python
#rename columns in first merger
combined_dataframe1.rename(columns={'Country': 'Country', 'Happiness Rank_x': 'Rank 2015', 'Happiness Score_x': 'Score 2015', 'Happiness Rank_y': 'Rank 2016', 'Happiness Score_y': 'Score 2016'}, inplace=True)
combined_dataframe1.head()
```

```python
# Check to see renamed columns in first merger
print(combined_dataframe1)
```

```python
# Combine the data into a single DataFrame again incorporating the next year, being 2017.
combined_dataframe2 = pd.merge(combined_dataframe1, happiness_2017,
                              how='left', on='Country')
```

```python
# check the merge
print(combined_dataframe2)
```

```python
#rename columns
combined_dataframe2.rename(columns={'Happiness.Rank': 'Rank 2017', 'Happiness.Score': 'Score 2017'}, inplace=True)
```

```python
#check the merged data frame after renaming columns
print(combined_dataframe2)
```

```python
print(combined_dataframe2.columns)
```

```python
# Combine the data into a single DataFrame again, but merging with 2018
combined_dataframe3 = pd.merge(combined_dataframe2, happiness_2018, 
                               how='left', on='Country')
```

```python
#check the column names of merged df
print(combined_dataframe3)
```

```python
combined_dataframe3.head()
```

```python
#rename columns
combined_dataframe3.rename(columns={'Overall rank': 'Rank 2018', 'Score': 'Score 2018'}, inplace=True)
```

```python
#check df of merged df
print(combined_dataframe3)
```

```python
# Combine the data into a single DataFrame again
happiness_index = pd.merge(combined_dataframe3, happiness_2019,
                              how='left', on='Country')
```

```python
#rename columns one last time
happiness_index.rename(columns={'Overall rank': 'Rank 2019', 'Score': 'Score 2019'}, inplace=True)
```

```python
#check the first five rows of fully merged df
happiness_index.head()
```

```python
#print the first five rows again
happiness_index.head()
```

```python
#replace na values with 0
happiness_index.fillna(0, inplace=True)
```

```python
happiness_index['Rank 2016'] = happiness_index['Rank 2016'].astype(int)
happiness_index['Rank 2017'] = happiness_index['Rank 2017'].astype(int)
happiness_index['Rank 2018'] = happiness_index['Rank 2018'].astype(int)
happiness_index['Rank 2019'] = happiness_index['Rank 2019'].astype(int)
happiness_index.head()
```

```python
#look for na values or anything that stands out 
unique_values = happiness_index['Score 2017'].unique()
print(unique_values)
```

```python
happiness_index['Score 2017'] = happiness_index['Score 2017'].round(3)
```

```python
#export file into a cleaned dataset csv file
file_path = 'Cleaned_Datasets/cleaned_happiness_index.csv'
happiness_index.to_csv(file_path, index=False)
```
