```python
import pandas as pd
csv_file = r'Data_Sources/ED Visits for Asthma (per year and state in USA.(2015-2021).csv'

# Use pandas to read the CSV file into a DataFrame
emergency_df = pd.read_csv(csv_file)
emergency_df.head()
```

```python
#delete columns 'StateFIPS', 'Data Comment'. 'Unnamed:5'
emergency_df.drop(['StateFIPS', 'Data Comment', 'Unnamed: 5'], axis=1, inplace=True)
emergency_df.head()
```

```python
#rename Value column
emergency_df = emergency_df.rename(columns={'Value': 'Emergency Dept Visits'})
emergency_df.head()
```

```python
cleaned_emergency = 'cleaned_emergencyvisits.csv'

#export the DataFrame as a CSV file
emergency_df.to_csv(cleaned_emergency, index=False)
```
